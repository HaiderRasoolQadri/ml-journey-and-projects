```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
```

## Log Transformation

```python
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data.head()
```

```python
# Adding a small constant to avoid log(0) and division by zero
data_positive = data + 0.01  # This ensures all values are positive
data_positive.head()
```

```python
log_data = np.log(data_positive)
log_data.head()
```

## Square Transformation

```python
data.head()
```

```python
squared_data = data ** 2
squared_data.head()
```

## Square Root Transformation

```python
data.head()
```

```python
sqrt_data = np.sqrt(data)
sqrt_data.head()
```

## Reciprocal Transformation

```python
data.head()
```

```python
reciprocal_data = 1 / data
reciprocal_data.head()
```
