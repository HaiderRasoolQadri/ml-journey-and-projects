```python
import numpy as np
import pandas as pd
```

```python
# load the dataset 
df = pd.read_csv("cars.csv")
df.head()
```

## 1. OHE using Pandas

```python
df['fuel'].value_counts()
```

```python
# Lets encode fuel and owner
pd.get_dummies(df, columns = ['fuel', 'owner'])
```

```python
# Droping the first category 
pd.get_dummies(df, columns = ['fuel', 'owner'], drop_first = True)
```

## 2. OHE using sklearn

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False) # To drop first column just add drop = 'first'

ohe.fit_transform(df[['fuel', 'owner']])
```

```python
# Converting into dataframe
ohe_df = pd.DataFrame(ohe.fit_transform(df[['fuel', 'owner']]), columns = ohe.get_feature_names_out(['fuel', 'owner']))
ohe_df.head()
```

```python
# Lets merge both dataframes
merged_df = pd.concat([df, ohe_df], axis = 1)
merged_df.head()
```
