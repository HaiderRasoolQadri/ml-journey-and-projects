# Feature Importance Calculation in Decision Trees

In a **Decision Tree**, feature importance tells you **how much each feature contributes**
to reducing prediction error (or impurity) when the tree makes splits.

Each split in the tree **reduces impurity** — e.g., Gini, entropy, or MSE — depending on the type of model:

- **Classification tree:** uses **Gini impurity** or **entropy**  
- **Regression tree:** uses **MSE (Mean Squared Error)**  

The **importance** of a feature is based on **how much total impurity it reduces** across the entire tree.


##  Formula (General Form)

For every node $(j)$ in the tree:

$$
\text{Importance of node } j = w_j \times \Delta I_j
$$

where:

- $(w_j)$ = number of samples reaching node $(j)$
- $\Delta I_j = I_{parent} - (p_L I_L + p_R I_R)$
  - $I_{parent}$ = impurity before split  
  - $I_L, I_R$ = impurity of left and right child nodes  
  - $p_L, p_R$ = proportion of samples that go left/right

Then for each **feature \( f \)**:

$$
\text{Feature importance}(f) =
\sum_{j \, \text{where split uses } f}
w_j \times \Delta I_j
$$

Finally, normalize so that importances sum to 1:

$$
\text{Normalized importance}(f) =
\frac{\text{Feature importance}(f)}
{\sum_k \text{Feature importance}(k)}
$$


```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
X,y = make_classification(n_samples=5, n_classes=2,
                               n_features=2, n_informative=2, n_redundant=0,
                               random_state=0)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X,y)
```

```python
plt.figure(figsize=(8, 10))
plot_tree(clf)
plt.show()
```

```python
clf.feature_importances_
```
