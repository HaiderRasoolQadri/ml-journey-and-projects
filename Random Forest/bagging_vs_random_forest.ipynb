## Differences between Bagging and Random Forest

### 1. Feature Selection

- Bagging: Uses all features when splitting a node in each tree.
- Random Forest: Uses only a random subset of features at each split.

### 2. Type of Base Model

- Bagging: Can use any model (like decision trees, SVMs, etc.).
- Random Forest: Uses only decision trees as base models.

### 3. Correlation Between Trees

- Bagging: Trees can become similar (highly correlated) since all features are used.
- Random Forest: Trees are less correlated because each considers different random features.

### 4. Overfitting

- Bagging: Reduces variance but might still overfit if trees are similar.
- Random Forest: Further reduces overfitting due to added randomness in feature selection.

### 5. Accuracy and Performance

- Bagging: Improves stability and accuracy over a single model.
- Random Forest: Usually gives better accuracy and generalization than bagging.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.tree import plot_tree
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X,columns=['feature1','feature2','feature3','feature4','feature5'])
df['target'] = y
df.head()
```

```python
print(df.shape)
```

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,10))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rf = RandomForestClassifier(max_features=2)
```

```python
rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,10))
plot_tree(rf.estimators_[4])
plt.show()
```
