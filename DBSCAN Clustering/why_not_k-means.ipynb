```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
```

```python
# Create concentric circle dataset
X, _ = make_circles(n_samples=500, factor=0.5, noise=0.03, random_state=4)
```

```python
# Applying K-MEANS
kmeans = KMeans(n_clusters=2, random_state=4)
kmeans_labels = kmeans.fit_predict(X)
```

```python
# Applying  DBSCAN
dbscan = DBSCAN(eps=0.1, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
```

```python
# Plot K-Means 
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', marker='o')
plt.title("K-Means Clustering (n_clusters=2)")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")

# Plot DBSCAN 
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering (eps=0.1, min_samples=5)")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")

plt.tight_layout()
plt.show()
```

**Interpretation**

**K-Means Result**

- K-Means divides the space into two circular boundaries, but because it assumes clusters are spherical and convex, it will cut through the rings incorrectly.
- K-Means minimizes the distance to the cluster centroids, so it fails when clusters have non-linear boundaries.
- The algorithm will likely split the two rings vertically or diagonally, mixing inner and outer circle points  poor clustering result.

**DBSCAN Result**

- DBSCAN identifies each ring as a separate cluster because it groups points based on density (not shape).
- Points far from both rings (if any) would be labeled as noise (-1).
- DBSCAN looks for dense regions separated by low-density gaps  perfect for concentric circle patterns.
- DBSCAN successfully recovers two distinct circular clusters  excellent performance here.
