```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=25)
plt.show()
```

## Step Function

```python
# Step function
def step(z):
    return 1 if z > 0 else 0
```

```python
def perceptron(X, y):
    X = np.insert(X, 0, 1, axis=1)
    # Initializing parameters and weight
    weights = np.ones(X.shape[1])
    lr = 0.01

    for i in range(1000):
        j = np.random.randint(0, 100) # Pick a random observation
        y_hat = step(np.dot(X[j], weights))
        weights = weights + lr * (y[j] - y_hat) * X[j]

    return weights[0], weights[1:]
```

```python
intercept_, coef_ = perceptron(X, y)
```

```python
intercept_, coef_
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(12,6))
plt.plot(x_input,y_input,color='red',linewidth=2, label='Perceptron Line')
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis',s=25)
plt.ylim(-3,2)
plt.legend()
plt.show()
```

## Sigmoid Function

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = sigmoid(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input2 = np.linspace(-3,3,100)
y_input2 = m*x_input + b
```

```python
plt.figure(figsize=(12,6))
plt.plot(x_input,y_input,color='red',linewidth=2, label='Perceptron Line')
plt.plot(x_input2,y_input2,color='brown',linewidth=2, label='Sigmoid Line')
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis',s=25)
plt.ylim(-3,2)
plt.legend()
plt.show()
```

## Logistic Regression

```python
log_r = LogisticRegression()

log_r.fit(X,y)
```

```python
m = -(log_r.coef_[0][0]/log_r.coef_[0][1])
b = -(log_r.intercept_/log_r.coef_[0][1])
```

```python
x_input3 = np.linspace(-3,3,100)
y_input3 = m*x_input + b
```
