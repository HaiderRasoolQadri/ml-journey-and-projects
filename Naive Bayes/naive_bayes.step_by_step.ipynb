# Training Phase

```python
# Calculate the probability of yes and no
df['play'].value_counts()
```

```python
P_yes = 9/14
P_no = 5/14

print(f'Probability of Yes is {P_yes} and No is {P_no}')
```

```python
# outlook

pd.crosstab(df['outlook'], df['play'])
```

```python
# p is probabaility
p_outlook_y = 4/9
p_rain_y = 3/9
p_sunny_y = 2/9

p_outlook_n = 0
p_rain_n = 2/5
p_sunny_n = 3/5
```

```python
# temp
pd.crosstab(df['temp'], df['play'])
```

```python
p_cool_y = 3/9
p_hot_y= 2/9
p_mild_y= 4/9

p_cool_n = 1/5
p_hot_n = 2/5
p_mild_n = 2/5
```

```python
# humidity
pd.crosstab(df['humidity'], df['play'])
```

```python
p_high_y = 3/9
p_normal_y = 6/9

p_high_n = 4/5
p_normal_n = 1/5
```

```python
# wind
pd.crosstab(df['wind'], df['play'])
```

```python
p_strong_y = 3/9
p_weak_y = 6/9

p_strong_n = 3/5
p_weak_n = 2/5
```

# Testing

```python
# From Problem 1
# Outlook-Sunny, temp-Hot, humidity-High, wind-Weak, play-No,P(Yes)

# Solution
# P(yes/sunny,hot,high,weak) = P(SunnyYes) * P(Hot/Yes) * P(Humidity/Yes) * P(Weak/Yes) * P(Yes)
# P(No/sunny, hot, high, weak) = P(Sunny/No) * P(Hot/No) * P(Humidty/No) * P(Weak/No) * P(No)

Pyes = p_sunny_y * p_hot_y * p_high_y * p_weak_y * P_yes
Pno = p_sunny_n * p_hot_n * p_high_n * p_weak_n * P_no
```

```python
print("Probability of Yes", np.round(Pyes, 2))
print("Probability of No", np.round(Pno, 2))

print("Final Result is `No` because Probability of No is Higher than Yes.")
```

```python
# Problem 2

# Outlook=Overcast , Temp-Mild, Humidity-Normal, wind= Strong
# Play or no play 

# Solution
# P(Yes/ Overcast, Mild, Normal,Strong) = P(Overcast/Yes) * P(Mild/Yes) * P(Normal/Yes) * P(Strong/Yes) *P(yes)
# P(No/Overcast, Mild, Normal,Strong) = P(Overcast/No) * P(Mild/No) * P(Normal/No) * P(Strong/No) * P(no)

pyes = P_yes * p_outlook_y * p_mild_y * p_normal_y * p_strong_y
pno = P_no * p_outlook_n * p_mild_n * p_normal_n * p_strong_n
```

```python
print("Probability of Yes", np.round(pyes, 2))
print("Probability of No", np.round(pno, 2))

print("Final Result is `Yes` because Probability of Yes is Higher than No.")
```
