# Applying Feature Construction

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1
```

```python
X.head()
```

```python
def familyCat(num):
    if num == 1: # alone
        return 0
    elif num > 1 and num <= 4: # Small Family
        return 1 
    else:
        # Large family
        return 2
```

```python
X['Family_type'] = X['Family_size'].apply(familyCat)
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Family_size'], axis=1, inplace=True)
X.head()
```
